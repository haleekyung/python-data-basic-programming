```python
# 원하는 영역에서 하이퍼링크 모두 추출하기
news = 'https://news.daum.net/'
soup = bs(ur.urlopen(news).read(), 'html.parser')
```

```python
for i in soup.find_all('div', {"class" : "item_issue"}):
    print(i.find_all('a'))
```

```python
# get으로 href 속성값 구하기
for i in soup.find_all('div', {"class" : "item_issue"}):
    print(i.find_all('a').get('href'))

# 바로 get을 사용하면 오류 메세지 뜸 : find_all의 값이 리스트이기 때문
# get은 특정 태그에서 속성값을 가져오는 것, find_all은 결과값으로 여러 태그가 들어가 있는 리스트를 출력하기 때문에 이 결괏값에 바로 get을 사용하면 제대로 작동안됨.
# find_all, text 작성시에도 마찬가지
```

```python

for i in soup.find_all('div', {"class" : "item_issue"}):
    print(i.find_all('a')[0].get('href'))
```

```python
# find_all과 get의 차이
# find_all : 태그를 추출할 때 사용함
# get : 태그 안에 있는 속성값을 추출할 때 사용함
```

```python
# 기사의 제목과 내용 한꺼번에 추출하기
# 웹 문서를 뷰티풀수프 객체에다 저장

article = 'https://go.seoul.co.kr/news/newsView.php?id=20200427004004&wlog_tag3=daum'
```

```python
# 기사 제목 가져오기
headline = soup.find_all('div', {'class' : 'item_issue'})
```

```python
print(headline[0].text)
```

```python
# 하이퍼링크된 모든 기사의 제목과 본문 추출하기
# headline 객체에서 모든 원소의 텍스트를 출력. for문 사용하기

for i in headline:
    print(i.text, '\n')
```

```python
for i in headline:
    print(i.text, '\n')
    soup3 = bs(ur.urlopen(i.find_all('a')[0].get('href')).read(), 'html.parser')
    for j in soup3.find_all('p'):
        print(j.text)
```
